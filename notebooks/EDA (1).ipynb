```python
plt.figure(figsize=(6,4))
sns.countplot(x=df["Class"])
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df["Class"]==1]["Time"], bins=50, color="red", label="Fraud", kde=True)
sns.histplot(df[df["Class"]==0]["Time"], bins=50, color="blue", label="Non-Fraud", kde=True)
plt.legend()
plt.title("Transaction Time Analysis")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Class'], y=df['Amount'], palette=['blue', 'red'])
plt.title("Transaction Amount by Class")
plt.xlabel("Class (0 = Legit, 1 = Fraud)")
plt.ylabel("Amount ($)")
plt.yscale("log")  # Log scale to handle skewness
plt.show()
```
